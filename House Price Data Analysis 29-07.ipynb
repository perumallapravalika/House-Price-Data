```python
#put cont. variables and cat. variables in 2 different lists

con_cols=[]
cat_cols=[]
for col in hp_df.columns:
    if hp_df[col].dtype=='int64' or hp_df[col].dtype=='float64':
        con_cols.append(col)   #to get values into the empty list
    if hp_df[col].dtype=='object':
        cat_cols.append(col)
con_cols
```

```python
#List Comprehension

con_cols=[col for col in hp_df.columns if hp_df[col].dtype=='int64' or hp_df[col].dtype=='float64']
cat_cols=[col for col in hp_df.columns if hp_df[col].dtype=='object']

print(con_cols)
```
